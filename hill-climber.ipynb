## Task 1: Hill-Climber

```python
# all imports
import numpy as np
from utils.objective_functions import griewank_objective_function
```

```python
from utils.plotter import Plotter
from algorithms.hill_climber.hill_climber import hill_climber


dimension = 2  # Change to 2 for 2D problems.
x0 = np.random.uniform(-10, 10, dimension)

best_solution, best_value, best_values_history, accepted_positions = hill_climber(
    griewank_objective_function, x0, num_iterations=5000, step_size=0.5
)

print("\n2D Problem:")
print("Best solution found:", best_solution)
print("Objective value:", best_value)

# Instantiate the plotter.
plotter = Plotter(ylabel="Objective Value")
# Plot the objective value history (works for any dimension).
plotter.plot_line(best_values_history)
# If the problem is 2D, plot the 2D movement path.
plotter.plot_path(accepted_positions, best_solution)

# --- Example 2: Solving an n-Dimensional problem (n != 2) ---
dimension = 5  # For example, a 5-dimensional problem.
x0_nd = np.random.uniform(-10, 10, dimension)

best_solution_nd, best_value_nd, best_values_history_nd, accepted_positions_nd = hill_climber(
    griewank_objective_function, x0_nd, num_iterations=5000, step_size=0.5
)

print("\n5D Problem:")
print("Best solution found:", best_solution_nd)
print("Objective value:", best_value_nd)

# For non-2D problems, we only plot the objective value history.
plotter_nd = Plotter(title="5D Hill-Climber Optimization Progress")
plotter_nd.plot_line(best_values_history_nd)
# Attempting to plot a 2D path will trigger a warning:
plotter_nd.plot_path(accepted_positions_nd, best_solution_nd)
```
